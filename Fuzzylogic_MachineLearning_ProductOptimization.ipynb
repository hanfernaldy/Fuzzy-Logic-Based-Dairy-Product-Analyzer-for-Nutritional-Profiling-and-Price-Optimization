```python
#Library
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("dataset_nutrisi_susu.xlsx")
df
```

```python
#Define data
sample = df["sample number"]
price  = df["price (Rp.)"]
nutrition = df["total nutrition (g)"]
```

```python
y = []

for x in range(31):
  Pvalue = [0, 0, 0]
  Nvalue = [0, 0, 0]
  bad = medium = good = expensive = mid = cheap = 0

  #Fuzzyfication
  if price[x] >= 8000:
      expensive = 1
      Pvalue[0] = expensive

  elif price[x] > 5000 and price[x] < 8500:
      expensive = (8500 - price[x]) / 500
      mid = (price[x] - 5000) / 500
      Pvalue[0] = expensive
      Pvalue[1] = mid

  elif price[x] >= 5500 and price[x] <= 8000:
      mid = 1
      Pvalue[1] = mid

  elif price[x] > 5000 and price[x] < 5500:
      mid = (5500 - price[x]) / 500
      cheap = (price[x] - 5000) / 500
      Pvalue[1] = mid
      Pvalue[2] = cheap

  elif price[x] <= 5000:
      cheap = 1
      Pvalue[2] = cheap

  if nutrition[x] <= 20:
      bad = 1
      Nvalue[0] = bad

  elif nutrition[x] > 20 and nutrition[x] < 25:
      bad = (25 - nutrition[x]) / 5
      medium = (nutrition[x] - 20) / 5
      Nvalue[0] = bad
      Nvalue[1] = medium

  elif nutrition[x] >= 25 and nutrition[x] <= 30:
      medium = 1
      Nvalue[1] = medium

  elif nutrition[x] > 30 and nutrition[x] < 35:
      medium = (35 - nutrition[x]) / 5
      good = (nutrition[x] - 30) / 5
      Nvalue[1] = medium
      Nvalue[2] = good

  elif nutrition[x] >=35 and nutrition[x] <= 50:
      good = 1
      Nvalue[2] = good

  #Inference
  NR = []
  if Nvalue[0] == bad and Pvalue[0] == expensive :
      NR.append(min(Pvalue[0], Nvalue[0]))
  if Nvalue[1] == medium and Pvalue[2] == cheap :
      NR.append(min(Pvalue[2], Nvalue[1]))
  if Nvalue[0] == bad and Pvalue[1] == mid :
      NR.append(min(Pvalue[1], Nvalue[0]))

  NRvalue = max(NR)

  R = []
  if Nvalue[2] == good and Pvalue[0] == expensive :
      R.append(min(Pvalue[0], Nvalue[2]))
  if Nvalue[1] == medium and Pvalue[1] == mid :
      R.append(min(Pvalue[1], Nvalue[1]))
  if Nvalue[0] == bad and Pvalue[2] == cheap :
      R.append(min(Pvalue[2], Nvalue[0]))

  Rvalue = max(R)

  HR = []
  if Nvalue[2] == good and Pvalue[1] == mid :
      HR.append(min(Pvalue[1], Nvalue[2]))
  if Nvalue[2] == good and Pvalue[2] == cheap :
      HR.append(min(Pvalue[2], Nvalue[2]))
  if Nvalue[1] == medium and Pvalue[2] == cheap :
      HR.append(min(Pvalue[2], Nvalue[1]))

  HRvalue = max(HR)

  #Defuzzification
  divide = NRvalue+Rvalue+HRvalue
  z = ((NRvalue*25)+(Rvalue*50)+(HRvalue*100)) / divide

  if z <= 100 and z > 50 :
    decision = "*Highly Recommended*"
  elif z <= 50 and z > 25 :
    decision = "*Recommended*"
  elif z <= 25 and z > 0 :
    decision = "*Not Recommended*"

  y.append([sample[x], z, decision])
```

```python
df['membership'] = y
print(df)
```
